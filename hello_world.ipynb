## 4. Using Variables and String Formatting

Let's store "Hello World" in a variable and display it in different ways.

```python
# Store message in a variable
message = "Hello World"

# Method 1: Simple print
print(message)

# Method 2: f-string formatting
print(f"Message: {message}")

# Method 3: String concatenation
print("The message is: " + message)

# Method 4: Multiple times
print(message * 3)

# Method 5: Uppercase
print(message.upper())
```
